```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
                   'var1': [10,23,33,22,11,99,76,84,45],
                   'var2': [100,253,333,262,111,969,405,578,760]})
df6
```

```python
df6.groupby("groups").aggregate([np.min, np.median, np.max])
```

```python
df6.groupby("groups").aggregate([min, len, np.max])
```

```python
df6.groupby("groups").aggregate({"var1":(min, max), "var2":"median"})
```

```python
def filter_func(x):
  return x["var1"].mean()>40
df6.groupby("groups").filter(filter_func)
```

```python
df6.groupby("groups").filter(lambda x: x["var2"].sum()>100 )
```

```python
df_a = df6.iloc[:, 1:3]
df_a
```

```python
df_a.transform(lambda x:(x - x.mean()/x.std()))
```
