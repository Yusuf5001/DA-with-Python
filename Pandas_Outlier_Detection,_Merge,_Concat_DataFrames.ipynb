```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_data/weight-height.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
q1 = df.Height.quantile(0.25)
q3 = df.Height.quantile(0.75)
iqr = q3-q1
lower = q1 - 1.5*iqr
higher = q3 + 1.5*iqr

iqr, lower,higher
```

```python
df[(df.Height < lower) | (df.Height > higher)]
```

```python
no_outlier = df[(df.Height > lower)&(df.Height < higher)]
no_outlier.head()
```

```python
df1 = pd.DataFrame({ "ID":[1,2,3,5, 9],
                    "Col_1": [1,2,3,4,5],
                    "Col_2": [6,7,8,9,10],
                    "Col_3":[11,12,13,14,15],
                    "Col_4": ["apple", "orange", "banana", "strawberry", "raspberry"]
})


df2 = pd.DataFrame({ "ID":[1,1,3,5],
                    "Col_A":[8,9,10,11],
                    "Col_B":[12, 13,15,17],
                    "Col_4":["apple", "orange", "banana", "kiwi"],
                      #"Col_4":[1,2,3,4]
    
})
```

```python
df1
```

```python
df2
```

```python
inner = pd.merge(df1, df2)
inner
```

```python
pd.merge(df1, df2, on="ID")
```

```python
pd.merge(df1, df2, on=["ID", "Col_4"])
```

```python
pd.merge(df1, df2, suffixes=["_l", "_r"], left_on="Col_2", right_on="Col_A")
```

```python
pd.merge(df1, df2, suffixes=["_l", "_r"], left_index=True, right_index=True)
```

```python
pd.merge(df1, df2, on="Col_4", how='outer')
```
