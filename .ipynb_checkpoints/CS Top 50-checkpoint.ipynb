## Data Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
```

```python
plt.figure(figsize=(9,6))

sns.scatterplot(x="Maps", y="Rounds", data=df);
plt.show()
```

```python
sns.displot(x= "Rating2.0", data=df, kde=True)
plt.show()
```

```python
sns.histplot(data=df, x="New_Rating", bins=10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

df["Rating2.0"].sort_values(ascending=False)[:5].plot(kind='bar', ax=ax)
ax.set_xticklabels(df.sort_values(by="Rating2.0", ascending=False)["Player"][:5], rotation=45)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+10));
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

df["New_Rating"].sort_values(ascending=False)[:5].plot(kind='bar', ax=ax)
ax.set_xticklabels(df.sort_values(by="New_Rating", ascending=False)["Player"][:5], rotation=45)


for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+10));
    
plt.show()
```
