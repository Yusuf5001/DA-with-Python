## Data Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
```

```python
plt.figure(figsize=(9,6))

sns.scatterplot(x="Maps", y="Rounds", data=df);
plt.show()
```

```python
sns.displot(x= "Rating2.0", data=df, kde=True)
plt.show()
```

```python
sns.histplot(data=df, x="New_Rating", bins=10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax = df["Rating2.0"].sort_values(ascending=False)[:5].plot(kind='bar', ax=ax)
ax.set_xticklabels(df.sort_values(by="Rating2.0", ascending=False)["Player"][:5], rotation=45)

for p in ax.patches:
     ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.25, p.get_height()), ha='center', va='top', color='yellow', size=18)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

df["New_Rating"].sort_values(ascending=False)[:5].plot(kind='bar', ax=ax)
ax.set_xticklabels(df.sort_values(by="New_Rating", ascending=False)["Player"][:5], rotation=45)


for p in ax.patches:
     ax.annotate(f'\n{p.get_height()+0.1}', (p.get_x()+0.02, p.get_height()), size=14)
plt.show()
```

```python
# with pandas
fig, ax = plt.subplots(figsize=(15,8))
df.sort_values(by="K-D Diff", ascending=False)["K-D Diff"][:10].plot(rot=45, kind='bar', ax=ax)

ax.set_ylabel("K/D Diff")
ax.set_xlabel("Player")
ax.set_title("Top 10 Players with highest K/D Difference")
ax.set_xticklabels(df.sort_values(by="K-D Diff", ascending=False)["Player"][:10], rotation=45)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.10, p.get_height()+0.5), size=15);
```

```python
fig, ax = plt.subplots(figsize=(15,8))
df.sort_values(by="K/D", ascending=False)["K/D"][:10].plot(rot=45, kind='bar', ax=ax)

ax.set_ylabel("K/D Ratio")
ax.set_xlabel("Player")
ax.set_title("Top 10 Players with highest K/D Ratio")
ax.set_xticklabels(df.sort_values(by="K/D", ascending=False)["Player"][:10], rotation=45)

for p in ax.patches:
     ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.25, p.get_height()), ha='center', va='top', color='yellow', size=18)
plt.show()
```

```python
# with pandas
fig, ax = plt.subplots(figsize=(15,8))
df.sort_values(by="K/D")["K/D"][:10].plot(rot=45, kind='bar', ax=ax)

ax.set_ylabel("K/D Ratio")
ax.set_xlabel("Player")
ax.set_title("Worst Players with highest K/D Ratio")
ax.set_xticklabels(df.sort_values(by="K/D")["Player"][:10], rotation=45)

for p in ax.patches:
     ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.25, p.get_height()), ha='center', va='top', color='yellow', size=18)
plt.show()
```
