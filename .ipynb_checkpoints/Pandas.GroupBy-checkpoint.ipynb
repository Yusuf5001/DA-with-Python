```python
# Index Levels
outside = ['M1', 'M1', 'M1', 'M2', 'M2', 'M2','M3', 'M3', 'M3']
inside = [1, 2, 3, 1, 2, 3, 5, 6, 7]
multi_index = list(zip(outside, inside))
multi_index
```

```python
hier_index = pd.MultiIndex.from_tuples(multi_index)
```

```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
                   'var1': [10, 23, 33, 22, 11, 99, 76, 84, 45],
                   'var2': [100, 253, 333, 262, 111, 969, 405, 578, 760]})
df6
```

```python
df6.agg(["min", "max"])
```

```python
df6[["var1", "var2"]].agg(["max", "min"])
```

```python
df6.agg({"var1":[sum],
        "var2":[min]
        })
```

```python
df6.agg({"var1":["sum", np.mean],
        "var2":["min", "max"]
        })
```

```python
df6.groupby("groups").agg([np.mean, max, np.min]).T
```

```python
df6.groupby("groups").agg({"var1":[np.mean,max],
                        "var2":["median", "min"]
                          }).T
```

```python
df6.filter(["groups", "var1"])
```

```python
df6[["groups", "var1"]]
```
