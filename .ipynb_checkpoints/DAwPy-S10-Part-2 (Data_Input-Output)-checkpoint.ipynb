## CSV

### CSV Input

```python
df = pd.read_csv("example.csv")
df
```

```python
pd.read_csv("ornekcsv.csv") # the problem is ; coz sep="," default
```

```python
pd.read_csv("ornekcsv.csv", sep=";") # use sep=; and problem solved
```

```python
pd.read_csv("ornekcsv.csv", sep=";", index_col=0) # will return a column as index 
```

```python
pd.read_csv("ornekcsv.csv", sep=";", index_col=2)
```

```python
pd.read_csv("ornekcsv.csv", sep=";", usecols=["a", "b"], nrows=5 ) # return only a and b columns and 5 rows
```

```python
pd.read_csv("ornekcsv.csv", sep=";", header=1)
```

```python
pd.read_csv("titanic.csv", sep="\t")
```

## Excel
Pandas can read and write excel files, keep in mind, this only imports data. Not formulas or images, having images or macros may cause this read_excel method to crash. 

### Excel Input

```python
df1 = pd.read_excel("Excel_Sample.xlsx") # will return sheet1, first page 
df1
```

```python
df2 = pd.read_excel("Excel_Sample.xlsx", sheet_name="Sheet2" )
df2
```

```python
df3 = pd.read_excel("Excel_Sample.xlsx", sheet_name=None) # will return sheets as dict
df3
```

```python
type(df3)
```

```python
pd.ExcelFile("Excel_Sample.xlsx").sheet_names # return names of the excel sheets
```
