```python
df
```

```python
df6 = df.fillna(method="pad", axis=1) # filling nan with previous on same row
df6.head()
```

```python
df7 = df.fillna(method="bfill", axis=1) # fillin nan on rows with next 
df7.head()
```

```python
df8 = df.fillna({"society":"abcd", "balcony":"defg"}) # filling nan of different columns with different values
df8.head(10)
```

```python
df9 = df.fillna(value=df["balcony"].mean()) # filling nans with the value of average of balcony
df9.head(10)
```

```python
df10 = df.fillna(value=df["balcony"].max()) #fill nan with value of max of balcony column
df10.head(10)
```

```python
df.head(10)
```

```python
df11 = df.dropna() # drop default nan
df11.head(10)
```

```python
df12 = df.dropna(how="all") # dropping row if all the rows are nan
df12
```

```python
df13 = df.dropna(how="any") # dropping row if any of row is nan
df13.head(10)
```

```python
df14 = df.replace(to_replace=np.nan, value=765) #replacing nan values
df14.head(10)
```

```python
df15 = df.replace(to_replace=3, value=5.0) #replacing 3 with 5
df15.head(10)
```

```python
df["balcony"] = df["balcony"].interpolate(method="linear")
df.head(10)
```

```python
df["balcony"] = df["balcony"].interpolate(method="linear", limit_direction="forward")
df.head(10)
```
